Se limpia el dataframe

```python
import pandas as pd
import re
def limpiar_dataframe(df_final):
    
   
    
    # 3. Quitar columnas basura
    df_final = df_final.drop(columns=["Unnamed: 0"], errors="ignore")
    
    # 4. Quitar filas sin texto
    if "tweet_text" in df_final.columns:
        df_final = df_final.dropna(subset=["tweet_text"])
    
    # 5. Limpiar texto (tweets)
    if "tweet_text" in df_final.columns:
        df_final["tweet_text"] = df_final["tweet_text"].astype(str).apply(
            lambda x: re.sub(r"http\S+|@\S+|#\S+", "", x).strip().lower()
        )
    
    # 6. Quitar duplicados
    df_final = df_final.drop_duplicates(subset="tweet_text")
    
    # 7. Reset index
    df_final = df_final.reset_index(drop=True)
    
    return df_final
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

nltk.download("vader_lexicon")

sia = SentimentIntensityAnalyzer()
```

Aplicar al dataframe 

```python
df_final["sentiment_score"] = df_final["tweet_text"].astype(str).apply(
    lambda x: sia.polarity_scores(x)["compound"]
)
```

```python
def label_sentiment(score):
    if score >= 0.05:
        return "positive"
    elif score <= -0.05:
        return "negative"
    else:
        return "neutral"

df_final["sentiment"] = df_final["sentiment_score"].apply(label_sentiment)
```
